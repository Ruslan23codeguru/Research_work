```python
def openfile (way):
    x = []
    y = []
    with open (way, 'r' ) as file:
        for i, line in enumerate(file):
            if i > 4:
                currentline = line.split(",")
                x.append(float(currentline[0]))
                y.append(float(currentline[1]))
    X = np.array(x)
    y = np.array(y)
    
     
    return X,y
```

```python
def create_list(data):
    new_data = np.zeros((40005,2))
    j = 0
    for i in range(data.shape[0]):
        if i%10 == 0:
            new_data[j] = data[i]
            j += 1
    return new_data
```

```python
def gauss_distr(mu, sigmaSquare, x):
    from math import sqrt, pi, e
    return (1 / sqrt(2 * pi * sigmaSquare)) * e ** ((-0.5) * (x - mu) ** 2 / sigmaSquare)
```

```python
def print_func(x,y):
    for i in range(len(x)):
        print(i ,"x = {}, y = {}".format(x[i], y[i]))
```
